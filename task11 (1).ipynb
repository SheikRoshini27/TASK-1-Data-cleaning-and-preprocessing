```python
# Step 1: Import required libraries
import pandas as pd
import zipfile
import os
from google.colab import files

# Step 2: Upload ZIP file
uploaded = files.upload()

# Step 3: Extract ZIP file
zip_filename = list(uploaded.keys())[0]  # get uploaded file name
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall("extracted_files")

# Step 4: Locate the CSV file (assuming only one CSV in ZIP)
csv_path = ""
for root, dirs, files_in_dir in os.walk("extracted_files"):
    for file in files_in_dir:
        if file.endswith('.csv'):
            csv_path = os.path.join(root, file)

# Step 5: Load dataset
df = pd.read_csv(csv_path)

# Step 6: Clean the data
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.drop_duplicates(inplace=True)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', format="%B %d, %Y")
df.dropna(subset=['title', 'type'], inplace=True)
df['country'] = df['country'].str.strip().str.title()

# Step 7: Show missing values summary
print("Missing values after cleaning:\n", df.isnull().sum())

# Step 8: Save and download cleaned dataset
cleaned_file = "netflix_titles_cleaned.csv"
df.to_csv(cleaned_file, index=False)
files.download(cleaned_file)
```
